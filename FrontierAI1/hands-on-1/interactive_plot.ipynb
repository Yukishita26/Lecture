```python
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import display, clear_output
```

```python
import time
import numpy as np

class Uprint:
  def __init__(self):
    self.outputs = ""
  def print(self, *args):
    s = " ".join(str(a) for a in args)
    self.outputs += "\n" + s
  def output(self):
    print(self.outputs)

que1 = []
que2 = []
ran = []

plt.ion()
fig = plt.figure(figsize=(12,4))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
u = Uprint()

for i in range(100):
  ran += [i]
  que1 += [np.sin(i/180*np.pi*5)]
  que2 += [np.cos(i/180*np.pi*5)]
  time.sleep(0.1)
  #print(i, i**2)
  u.print(ran[-1], que1[-1], que2[-1])
  
  clear_output(wait = True)
  u.output()
  ax1.cla()
  ax2.cla()
  ax1.plot(ran, que1, marker="o")
  ax2.plot(ran, que2, marker="o")
  #plt.xlim(0,10)
  #plt.ylim(0,100)
  display(fig) 
```
